```python
##########################
##### PROCESS VIDEOS #####
##########################
import os
import sleap
import numpy as np

#
slp_dir = '/scratch/cm5635/cohort2_slp/'

# 
fnames = np.loadtxt(os.path.join(slp_dir,'cohort2_fnames.txt'))
models = np.loadtxt(os.path.join(slp_dir,'cohort2_nn_types.txt'))

# randomly shuffle files 
idx = np.random.choice(np.arange(fnames.shape[0]), fnames.shape[0], replace=False)
for id_ in idx:

    # 
    fname = fnames[id_]

    ###############################################
    ##### CHECK IF FILE PROCESSED AND COPY IT #####
    ###############################################
    # check if file already done;
    fname_dummy = os.path.join(slp_dir,fname+"_in_process.txt")
    if os.path.exists(fname_dummy):
        print (" video already processed, skipping ...")
        continue
    else:
        # save a dummy variable that stores the model type and 
        np.savetxt(fname_dummy, [models[id_]],  delimiter=" ", fmt="%s")

        # copy the file from the cloud to local slp directory
        #  CAN SKIP THIS STEP IF TESTING THE RUN LOCALLY
        # just put the video in the slp_dir above and comment out this code
        command_txt = 'rclone copy --drive-shared-with-me mitelutco:cohort2_compressed_video/'+ fname + ' '+slp_dir

    ###############################################
    ##### OPTIONAL CODE TO COMPRESS/CROP VIS ######
    ###############################################
    # LISA TO INSERT CODE HERE
    # load video
    fname_vid = os.path.join(slp_dir,fname)

    # Load vid, shrhink it and cropp it
    #
    
    # save the final shirnk compressed video
    fname_shrink_compressed = os.path.join(slp_dir,
                                           fname.replace(".mp4","_shrink_cropped.mp4"))   
        
    ###############################################
    ############### GET MODEL NAMES ###############
    ###############################################
    fname_models = []
    model_names = []
    if models[id_]=='Night':
        fname_models.append(os.path.join(slp_dir, 'models','night','gerbils.multiclass_bottomup_night')
        model_names.append("Night")
    elif models[id_]=='Day':
        fname_models.append(os.path.join(slp_dir, 'models','night','gerbils.multiclass_bottomup_day')
        model_names.append("Day")
    elif models[id_]=='Both':
        fname_models.append(os.path.join(slp_dir, 'models','night','gerbils.multiclass_bottomup_night')
        fname_models.append(os.path.join(slp_dir, 'models','night','gerbils.multiclass_bottomup_day')
        model_names.append("Day")
        model_names.append("Night")
                               
    ###############################################
    ######## LOOP OVER EACH MODEL FOR VIDEO #######
    ###############################################
    #
    for ctr,fname_model in enumerate(fname_models):
        print ("Loading model: ", fname_model)
        predictor = sleap.load_model([fname_model], batch_size=16)
        
        # copy the video to cohort1 directory for processing       
        fname_local = os.path.join(slp_dir,fname.replace(".mp4","_shrink_cropped.mp4"))

        # load video
        video = sleap.load_video(fname_local)

        # predict video
        labels_pr = predictor.predict(video)

        # save labels
        fname_slp = os.path.join(slp_dir,fname.replace('.mp4','_')+model_names[ctr]+".slp")
        labels_pr.save(fname_slp)

print ("DONE")    
    
```
