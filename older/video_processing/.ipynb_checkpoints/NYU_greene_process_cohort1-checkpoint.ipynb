```python
#############################################################
##### SAVE MODEL TYPE FOR EACH VIDEO IN COHORT 1 + ##########
###  CORRECT AM / PM TIMES FOR COHORT1 + DAYLIGHT SAVINGS ###
#############################################################
# 
root_dir = 'data/dan_lab/cohort_1_march_2/compressed_videos/compressed_videos/'

# 
filenames_ordered = '/media/cat/4TBSSD/dan/cohort1/filenames_reordered_clean.txt'
fnames = np.loadtxt(filenames_ordered, dtype='str')
print (fnames.shape)


# lights ON: 07:00;   lights OFF: 19:00 pre daylight savings  March 11, 2020
# lights ON: 06:00:   lights OFF: 18:00 post daylight savings

# 

pre_daylight_savings = True
am_pm = "PM"
fnames_fixed = []
models_fixed = []
daytime_fixed = []
day_ctr = 0
day_stacks = []
day_stacks.append([])
for fname in fnames:
    
    # 
    if fname == 'break':
        am_pm = "AM"
        continue
    elif fname == 'done':
        break
    
    # read day 
    vals = []
    for k in range(3):
        idx = fname.index('_')
        vals.append(fname[0:idx])
        fname = fname[idx+1:]
    day = int(vals[2])
    
    # find hour and convert TO 24 HR CLOCK
    hour = int(fname[:2])
    if am_pm=="PM" and hour != 12:
        hour += 12 
    elif hour == 12 and am_pm == 'AM':
        hour = 0
    
    # 
    minute = int(fname[3:5])
    
    # Correct the time shift 
    hour_shifted = hour-5
    
    # check if shifting back in time changed the time to negative
    am_pm_shifted = am_pm
    if hour_shifted < 0:
        hour_shifted+=24
        am_pm_shifted = "PM"
        day-=1

    if hour_shifted == 0:
        day_ctr+=1
        day_stacks.append([])
        print ("new day")
    
    # determin if day or night cycle
    #print (day, "dayligh savings", pre_daylight_savings, "  hour shifted: ", hour_shifted)
    if pre_daylight_savings and (hour_shifted >= 7 and hour_shifted < 19):
        model='daytime'
    elif pre_daylight_savings and (hour_shifted < 7 or hour_shifted >= 19):
        model='nighttime'
    elif not pre_daylight_savings and (hour_shifted >=6 and hour_shifted <18):
        model='daytime'
    elif not pre_daylight_savings and (hour_shifted <6 or hour_shifted >=18):
        model='nightime'
    else:
        print ("NO MODEL")
        break
        
    
    # 
    print ("day: ", day, "hour: ", hour_shifted, " min ",  minute, " using ", model)
    
    day_stacks[day_ctr].append([day, hour_shifted, minute, model])
    
    
    # FOR NEXT CYCLE
    if day == 11:
        pre_daylight_savings=False
    
    # switch to PM after time is 11AM <--- note this is not completely robust if there is a jump between 10:59am to 12:00+ PM 
    #                                     it will be missed
    if hour==11 and am_pm == "AM":
        am_pm = 'PM'        
    
    
    
    # 
    fnames_fixed.append(fname)
    models_fixed.append(model)
    daytime_fixed.append(["original", fname, " corrected ", day, hour_shifted, minute, model])
    
    
# resave the fname
np.savetxt('/home/cat/fnames_fixed.txt', fnames_fixed, 
               delimiter=" ", fmt="%s")
# save the model type
np.savetxt('/home/cat/models_fixed.txt', models_fixed, 
               delimiter=" ", fmt="%s")

# save the model type
np.savetxt('/home/cat/daytime_fixed.txt', daytime_fixed, 
               delimiter=" ", fmt="%s")

#np.savetxt('/home/cat/daystacks.txt', day_stacks, 
#               delimiter=",", fmt="%s")

np.save('/home/cat/daystacks.txt', day_stacks)
    
       
```
