```python
#################################################
####### SMOOTH TRACES # COMPUTE VELOCITIES ######
#################################################
# 
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

from tqdm import trange

fname_out = '/media/cat/4TBSSD/dan/cohort1/alldata.npz'
data = np.load(fname_out)
locs = data['locs']
day_times = data['day_times']
               
# smooth out trajectories using sliding window
box_width = 25
locs = np.array(locs)
print (locs.shape)
locs_smooth = locs.copy()
for k in trange(locs.shape[0]):
    for p in range(locs.shape[2]):
        for q in range(locs.shape[3]):
            locs_smooth[k,:,p,q] = smooth(locs_smooth[k,:,p,q],
                                    box_width)

# compute velocities for each animal
vel = []
for k in range(locs_smooth.shape[0]):
    temp = locs_smooth[k]
    vel.append([])
    for p in range(locs_smooth.shape[2]):
        temp_vel = np.sqrt((temp[1:,p,0]-temp[:-1,p,0])**2+
                           (temp[1:,p,1]-temp[:-1,p,1])**2)
        #print ("temp vel: ", temp_vel.shape)
    
        vel[k].append(temp_vel)
    
vel = np.array(vel)
print ("Velocities array: ", vel.shape)
```

```python
##############################################################
####### FIND MOVEMENT INITIATIONS AND WHO FOLLOWS WHO ########
##############################################################

# detect movement from any period of rest
frame_rate = 25
# min_rest_period = 5 * frame_rate

# walk vs. running thresholds
rest_velocity = 5  #very slowly moving / stationary threshold
                   # measured in pixels/frame movement; 
min_activity_duration = int(.5 * frame_rate)  # movement < n frames between rest will be discarded (i.e. counted as rest)

# 
run_velocity = 20
min_run_duration = int(0.5*frame_rate)


# detect movement initiations from rest
mov_initiations = np.zeros((vel.shape[0],
                            vel.shape[1],
                            vel.shape[2]),'float32')

n_mov_initiations = np.zeros((vel.shape[0],
                              vel.shape[1]),'float32')

#
#rest_array = np.zeros((vel.shape[0],
#                       vel.shape[1]),'float32')

# keep track of distribution of resting periods;
activity_array = []

n_runs = np.zeros((vel.shape[0],
                              vel.shape[1]),'float32')

#
run_array = []

# loop over each hour
for k in trange(vel.shape[0]):
    
    activity_array.append([])
    run_array.append([])
        
    # loop over each animal:
    for a in range(vel.shape[1]):
        
        
        #########################################
        ######### FIND REST PERIODS #############
        #########################################
        
        activity_array[k].append([])
        run_array[k].append([])
        
        temp = vel[k,a]
        idx = np.where(np.isnan(temp))[0]
        temp[idx]=0
        
        # find periods where animals were at rest;
        idx = np.where(temp>=rest_velocity)[0]
        
        # 
        if idx.shape[0]<=1:
            continue
            
        # compute diffs
        diff = idx[1:]-idx[:-1]
        idx2 = np.where(diff>=min_activity_duration)[0]
        idx_starting  = idx[idx2]+1
        idx_ending = idx[idx2+1]-1
        
        # add the # of initiations
        n_mov_initiations[k,a] += idx_starting.shape[0]
#         print ("movs: ", idx_starting.shape)

#         print (idx_starting [:10], idx_ending[:10])
#         print (temp[idx_starting[0]-2:idx_starting[0]+2])
#         print (temp[idx_ending[0]-2:idx_ending[0]+2])
        
        # search for sequential periods > min duration as indicative of rest-> movement transitions
        activity_array[k][a].append([idx_ending-idx_starting])
        
        
        #########################################
        ######### FIND RUN PERIODS #############
        #########################################
                
        temp = vel[k,a]
        idx = np.where(np.isnan(temp))[0]
        temp[idx]=0
        
        # find periods where animals were at rest;
        idx = np.where(temp>=run_velocity)[0]
        #print (idx.shape)
        # 
        if idx.shape[0]<=1:
            continue
            
        # compute diffs
        diff = idx[1:]-idx[:-1]
        idx2 = np.where(diff>=min_run_duration)[0]
        idx_starting = idx[idx2+1]-1
        idx_ending  = idx[idx2]+1
#         print ("runs: ", idx_starting.shape)
        
        try:
            if idx_ending[0]< idx_starting[0]:
                idx_ending = idx_ending[1:]
                idx_starting = idx_starting[:-1]
        except:
            pass
#         id_ = 4
#         try:
#             print (idx_starting [:10], idx_ending[:10])
#             print (temp[idx_starting[id_]-3:idx_starting[id_]+3])
#             print (temp[idx_ending[id_]-2:idx_ending[id_]+2])
#         except:
#             pass
        
        # add the # of initiations
        n_runs[k,a] += idx_starting.shape[0]
       
        # search for sequential periods > min duration as indicative of rest-> movement transitions
        run_array[k][a].append([idx_ending[:-1]-idx_starting[1:]])
#         print ('')
        
        
        
#     if k>10:
#         break
print (n_mov_initiations.shape)
#plt.show()
```

```python


import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize


def normal_pdf(x, mean, var):
    return np.exp(-(x - mean)**2 / (2*var))


# Generate the space in which the blobs will live
xmin, xmax, ymin, ymax = (0, 100, 0, 100)
n_bins = 100
xx = np.linspace(xmin, xmax, n_bins)
yy = np.linspace(ymin, ymax, n_bins)

# Generate the blobs. The range of the values is roughly -.0002 to .0002
means_high = [20, 50]
means_low = [50, 60]
var = [150, 200]

gauss_x_high = normal_pdf(xx, means_high[0], var[0])
gauss_y_high = normal_pdf(yy, means_high[1], var[0])

gauss_x_low = normal_pdf(xx, means_low[0], var[1])
gauss_y_low = normal_pdf(yy, means_low[1], var[1])

weights = (np.outer(gauss_y_high, gauss_x_high)
           - np.outer(gauss_y_low, gauss_x_low))

# We'll also create a grey background into which the pixels will fade
greys = np.full((*weights.shape, 3), 70, dtype=np.uint8)
print (greys.shape)
# First we'll plot these blobs using ``imshow`` without transparency.
vmax = np.abs(weights).max()
imshow_kwargs = {
    'vmax': vmax,
    'vmin': -vmax,
    'cmap': 'RdYlBu',
    'extent': (xmin, xmax, ymin, ymax),
}

fig, ax = plt.subplots()
#ax.imshow(greys)
ax.imshow(weights, **imshow_kwargs)
ax.set_axis_off()

```

```python
# Create an alpha channel of linearly increasing values moving to the right.
alphas = np.ones(weights.shape)
alphas[:, 30:] = np.linspace(1, 0, 70)

# Create the figure and image
# Note that the absolute values may be slightly different
fig, ax = plt.subplots()
ax.imshow(greys)
ax.imshow(weights, alpha=alphas, **imshow_kwargs)
ax.set_axis_off()
```
